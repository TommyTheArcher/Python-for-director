**Задание 5.** Наёдите 95-процентный доверительный интервал.

```python
mean = np.mean(deposits)
std_dev = np.std(deposits, ddof=1) 
n = len(deposits)  

confidence = 0.95 # уровень доверия 95% 
z = st.norm.ppf(1 - (1 - confidence) / 2)  # критическое значение для 95%

standard_error = std_dev / np.sqrt(n) # ошибка

margin = z * standard_error 
confidence_interval = (mean - margin, mean + margin)

print(f"среднее значение: {mean:.2f} тыс. рублей")
print(f"95%-ный доверительный интервал: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f}) тыс. рублей")
```

**Задание 7**. Пусть представленные данные - выборка, а вы хотите сделать вывод о генеральной совокупности (всех вкладах в России). Постройте 95% доверительный интервал среднего размера банковского вклада.

```python
mean = np.mean(deposits)
std_dev = np.std(deposits, ddof=1) 
n = len(deposits)  

confidence = 0.95
z = st.norm.ppf(1 - (1 - confidence) / 2) 

standard_error = std_dev / np.sqrt(n)

margin = z * standard_error
confidence_interval = (mean - margin, mean + margin)

print(f"среднее значение выборки: {mean:.2f} тыс. рублей")
print(f"95%-ный доверительный интервал среднего: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f}) тыс. рублей")
```

Предположим, что вы прочитали в газете, что средний размер депозита составляет 306 тысяч рублей и с помощью выборочных данных, которые у вас есть, вы хотите это проверить.

**Задание 8**. Сформулируйте нулевую и альтернативную гипотезы.

```python
H0 = "H0: средний размер депозита в совокупности равен 306 тыс. рублей "
H1 = "H1: средний размер депозита в совокупности не равен 306 тыс. рублей "
print("нулевая гипотеза:")
print(H0)
print("\nальтернативная гипотеза:")
print(H1)
```

**Задание 9**. Вычислите p-value.

```python
mean = np.mean(deposits)  
std_dev = np.std(deposits, ddof=1) 
n = len(deposits)  
mu_0 = 306  # предполагаемое среднее по H0

t_stat = (mean - mu_0) / (std_dev / np.sqrt(n))

p_value = 2 * (1 - st.t.cdf(abs(t_stat), df=n-1)) # вычисляем p-значение для двустороннего теста

print(f"среднее значение выборки: {mean:.2f} тыс. рублей")
print(f"t-статистика: {t_stat:.4f}")
print(f"p-значение: {p_value:.4f}")
```

**Задание 10**. При пороговом значении в 10%, сделайте вывод о возможности либо невозможности отвергнуть нулевую гипотезу.

```python
p_value = 0.0303  
alpha = 0.10 

if p_value <= alpha:
    conclusion = "отвергаем: средний размер депозита отличается от 306 тыс. рублей."
else:
    conclusion = "не можем отвергнуть: нет оснований считать, что средний размер депозита отличается от 306 тыс. рублей."
```

```python
# Вывод результата
print(f"p-значение: {p_value:.4f}")
print(f"Уровень значимости: {alpha}")
print(conclusion)
```
