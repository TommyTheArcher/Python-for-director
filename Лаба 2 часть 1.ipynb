### Провести касательную к графику функции у = х^2 + 2 в точке с абсциссой х0 = 2.

```python
import numpy as np
from sympy import *
from sympy.abc import x
import matplotlib.pyplot as plt
```

```python
def tangent(y,x0):
# Строит уравнение касательной к графику функции у(х) в точке с абсциссой х0 (х0;у0) - точка касания 
   
    y0 = y.subs(x, x0)
    x1 = x0 + 1
# Производная функции у при х=х0 
    
    k = diff(y, x).subs(x, x0)
    y1 = y0 + k
    

    return Line((x0,y0), (x1,y1))
```

```python
y=x**2 + 2
x0 = 2
# Метод equation() позволяет вывести общее уравнение прямой 
tangent(y,x0).equation()
```

### 2. Исследовать на экстремум функцию у = е^х - е^(2х)

```python
# Определяем функцию
y=exp(x)-exp(2*x)
# Определяем первую производную функции
diff(y)
```

```python
# Вычисляем значения функции в критических точках
solve(diff(y),x)
```

```python
# Определяем, является ли критическая точка максимумом или минимумом
# Найдем 2-е производные 
# (y,x,x)- другой способ нахождения производной
diff(y,x,2).subs(x,-log(2))
```

#### 5. Интервалы выпуклости и точки перегиба графика функции.




```python
# находим 2ую производную(Точка перегиба — это точка, в которой функция изменяет свою выпуклость, то есть переходит из выпуклой вверх в выпуклую вниз или наоборот.)
print(solve(diff(y,x,2),x))
#выпуклость
print(diff(y,x,2).subs(x,-log(4)+1))
print(diff(y,x,2).subs(x,-log(4)-1))
print(round(diff(y,x,2).subs(x,-log(4)+1),2))#2ая произв отриц, значит 1ая убывает, значит выпуклость вверх
print(round(diff(y,x,2).subs(x,-log(4)-1),2))#2ая произв полож, значит 1ая возрастает, значит выпуклость вниз
```
