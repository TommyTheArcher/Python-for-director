```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
Найти площадь фигуры, ограниченной линиями у = 2х, у = -х^2 + 7х - 6.
```

```python
x = sp.symbols('x')
f1 = 2 * x  
f2 = -x**2 + 7*x - 6  

intersections = sp.solve(f1 - f2, x)
print(f"точки пересечения: x = {intersections}")

a, b = intersections[0], intersections[1]
area = sp.integrate(f2 - f1, (x, a, b))
print(f"площадь: {area}")
```

```python
Количество потребляемой предприятием электроэнергии меняется в течение суток в зависимости от времени t со скоростью v(t) = 8+4sin((t+7)/4), 
где время t измеряется в часах. Найти суммарный расход электроэнергии за сутки
```

```python
def v(t):
    return 8 + 4 * np.sin((t + 7) / 4)

t = np.linspace(0, 24, 100)
v_vals = v(t)
plt.plot(t, v_vals, color='green', label='v(t) = 8 + 4 sin((t+7)/4)')
plt.title('скорость потребления электроэнергии за сутки')
plt.xlabel('время, часы')
plt.ylabel('скорость, кВт')
plt.grid(True)
plt.legend()
plt.show()

# суммарный расход (интеграл от 0 до 24)
energy, error = quad(v, 0, 24)

print(f"суммарный расход электроэнергии за сутки: {energy:.2f} кВт·ч")
print(f"погрешность: {error:.2e}")
```

```python
Найти объём продукции, произведённый за 5 лет, если функция Кобба-Дугласа имеет вид: F(t) = (1 + t)e3t
```

```python
def F(t):
    return (1 + t) * np.exp(3 * t)

t = np.linspace(0, 5, 100)
F_vals = F(t)
plt.plot(t, F_vals, color='blue', label='F(t) = (1 + t)e^(3t)')
plt.title('скорость производства продукции за 5 лет')
plt.xlabel('время, годы')
plt.ylabel('скорость, ед./год')
plt.grid(True)
plt.legend()
plt.show()

# объём продукции 
volume, error = quad(F, 0, 5)

print(f"объём продукции за 5 лет: {volume:.2f} ед.")
print(f"погрешность: {error:.2e}")
```
